```python
import pandas as pd
```

```python
data = pd.read_csv(u"./Trade_Logs_DERUN_GUOLIAN/成交记录_160302.csv", encoding='gb2312')
data.head()
```

```python
def judge(x):
    if u'买' in x:
        return 1
    elif u'卖' in x:
        return -1
    else:
        return 9999
```

```python
data[u'正负'] = data[u'买卖'].apply(judge)
data.head()
```

```python
data[u'hold'] = data[u'成交手数']*data[u'正负']
data.head()
```

```python
a = data.groupby(by=u'合约').sum()[[u'hold']]
a[u'合约'] = a.index
a
```

```python
hold1 = pd.read_csv(u'./Trade_Logs_DERUN_GUOLIAN/持仓记录_160301.csv', encoding='gb2312')
hold1[u'正负'] = hold1[u'买卖'].apply(judge)
hold1[u'hold'] = hold1[u'正负']*hold1[u'总持仓']
hold1 = hold1.groupby(u'合约').sum()
hold1[u'合约'] = hold1.index
hold1
```

```python
hold1_amount = hold1.merge(a, how='outer', on=u'合约')
hold1_amount = hold1_amount.fillna(value=0)
hold1_amount
```

```python
hold1_amount[u'amount'] = hold1_amount[u'hold_x']+hold1_amount[u'hold_y']
hold1_amount
```

```python
out = hold1_amount[[u'合约', u'amount']]
out.head()
```

```python
def judje2(x):
    if x >= 0:
        return u'买 '
    elif x< 0:
        return u' 卖'
    else:
        return u'NULL'
```

```python
out[u'买卖'] = out[u'amount'].apply(judje2)
out[u'总持仓'] = pd.np.abs(out[u'amount'])
out
```
